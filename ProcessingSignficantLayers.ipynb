```python
def mv_E2F(dir_name, save_dir):
    
    '''
        将E盘下面的E:\Resource\DataSet\MedicalImage\SignificantLayersMultiExpand拷贝只F盘，每个文件的命名方式是patientid_checkid_lesionid_sliceid_type_id
    '''
    dir_names = os.listdir(dir_name)
    for sub_dir_name in dir_names:
        split_res = sub_dir_name.split('_')
        srrid = split_res[0]
        lesionid = split_res[1]
        sliceid = split_res[2]
        typeid = split_res[3]
        typename = return_typename_byid(int(typeid))
        cur_path = glob(
            os.path.join('E:\\Resource\\DataSet\\MedicalImage', typename, srrid+'-*')
        )[0]
        cur_path = glob(
            os.path.join(cur_path, 'ART', '*.DCM')
        )[0]
        patientid = read_patientId(cur_path)
        basename = os.path.basename(cur_path)
        checkid = basename[basename.find('_')+1: basename.find('_', basename.find('_') + 1)]
        target_dir = os.path.join(save_dir, patientid + '_' + checkid + '_' + lesionid + '_' + sliceid + '_' + typeid)
        if not os.path.exists(target_dir):
            os.mkdir(target_dir)
        filenames = os.listdir(os.path.join(dir_name, sub_dir_name))
        for filename in filenames:
            filepath = os.path.join(dir_name, sub_dir_name, filename)
            shutil.copy(
                filepath,
                os.path.join(target_dir, filename)
            )
```

```python
def split_trainval(dir_name, save_dir):
    '''
        将数据集划分成训练和测试
    '''
    names = os.listdir(dir_name)
    record_dict = {}
    for name in names:
        if name in record_dict.keys():
            continue
        lesion_id = name[:name.find('_', name.find('_', name.find('_') + 1) + 1)]
        paths_withsamelesion = glob(os.path.join(dir_name, lesion_id + '*'))
        random_flag = np.random.rand()
        if random_flag > 0.5:
            for path in paths_withsamelesion:
                if os.path.exists(os.path.join(save_dir, 'train', os.path.basename(path))) or os.path.exists(os.path.join(save_dir, 'val', os.path.basename(path))):
                    continue
                shutil.copytree(
                    path,
                    os.path.join(save_dir, 'train', os.path.basename(path))
                )
                
                record_dict[os.path.basename(path)] = 1
        else:
            for path in paths_withsamelesion:
                if os.path.exists(os.path.join(save_dir, 'val', os.path.basename(path))) or os.path.exists(os.path.join(save_dir, 'train', os.path.basename(path))):
                    continue
                shutil.copytree(
                    path,
                    os.path.join(save_dir, 'val', os.path.basename(path))
                )
                record_dict[os.path.basename(path)] = 1
```

```python
def extract_roi(dir_name):
    def find_minrect(image, mask_image):
        xs, ys = np.where(mask_image != 0)
        xs_min = np.min(xs)
        xs_max = np.max(xs)
        ys_min = np.min(ys)
        ys_max = np.max(ys)
        return image[xs_min: xs_max+1, ys_min: ys_max+1]
    def compress22dim(image):
        shape = list(np.shape(image))
        if len(shape) == 3 and shape[0] == 1:
            return np.squeeze(image)
        return image
            
    '''
        提取每个显著层的ROI和ROIexpand，默认向外扩充10%
    '''
    for phasename in ['NC','ART', 'PV']:
        image_path = glob(os.path.join(dir_name, phasename+'_Image*.mhd'))[0]
        mask_path = glob(os.path.join(dir_name, phasename+'_Mask*.mhd'))[0]
        image = read_mhd_image(image_path)
        image[image < 0] = 0
        mask_image = read_mhd_image(mask_path)
        image = compress22dim(image)
        mask_image = compress22dim(mask_image)
        mask_image[mask_image != 1] = 0
        mask_image = fill_region(mask_image)
       
        print image_path, np.shape(image), np.shape(mask_image)
        if np.sum(mask_image) > 500*500:
            print 'Fill Error Happend', mask_path
            continue
        props = regionprops(mask_image)
        # expand_r = int(props[0].equivalent_diameter * 0.1)
        expand_r = 10
        # 判断中心区域是否已经填上了0
        centroid = props[0].centroid
        if mask_image[int(centroid[0]), int(centroid[1])] == 0:
            print 'Fill Failed', mask_path
            continue
        if expand_r < 5:
            expand_r = 5
        expand_mask_image = image_expand(mask_image, expand_r)
        expand_image = copy(image)
        expand_roi = expand_image
        expand_roi[expand_mask_image != 1] = 0
        roi = image
        roi[mask_image != 1] = 0
        roi = find_minrect(roi, mask_image)
        expand_roi = find_minrect(expand_roi, expand_mask_image)
        save_mhd_image(roi, os.path.join(dir_name, phasename+'_ROI.mhd'))
        save_mhd_image(expand_roi, os.path.join(dir_name, phasename+'_ROI_Expand.mhd'))
```

```python
def extractroi_all_dir(dir_name):
    '''
        提取dir_name文件夹下面的的所有子文件夹
    '''
    sub_dir_names = os.listdir(dir_name)
    for sub_dir_name in sub_dir_names:
        extract_roi(os.path.join(dir_name, sub_dir_name))
```

```python
def extractroi2target_dir(dir_name, target_dir):
    if not os.path.exists(target_dir):
        os.mkdir(target_dir)
    paths = glob(os.path.join(dir_name, '*ROI*'))
    if len(paths) != 12:
        print 'The number of roi is not equal 6, it is %d, in %s' % (len(paths), dir_name)
    for path in paths:
        shutil.copy(
            path,
            os.path.join(target_dir, os.path.basename(path))
        )
        if path.endswith('.mhd'):
            image = read_mhd_image(path)
            save_image(image, os.path.join(target_dir, os.path.basename(path).split('.mhd')[0] + '.png'))
```

```python
def extractroi2target_all_dir(dir_name, target_dir):
    sub_dir_names = os.listdir(dir_name)
    for sub_dir_name in sub_dir_names:
        extractroi2target_dir(os.path.join(dir_name, sub_dir_name), os.path.join(target_dir, sub_dir_name))
extractroi2target_all_dir('G:\\SL_TrainAndVal\\train', 'G:\\ROI\\train')
```
